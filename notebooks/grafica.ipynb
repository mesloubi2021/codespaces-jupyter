```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'t(s)': [0, 10, 20, 30, 40, 50],
        'x(m)': [30, 52, 38, 0, -37, -53]}

df = pd.DataFrame(data)
display(df)

plt.figure(figsize=(6, 4))
plt.plot(df['t(s)'], df['x(m)'], marker='o', color='red')
plt.xlabel('t(s)')
plt.ylabel('x(m)')
plt.title('Posición vs. Tiempo')
plt.grid(True)
plt.show()
```
