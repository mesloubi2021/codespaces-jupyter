```python
import matplotlib.pyplot as plt

# Datos
tiempo = [0, 10, 20, 30, 40, 50]
posicion = [30, 52, 38, 0, -37, -53]

# Crear el gráfico con color violeta
plt.figure(figsize=(8, 5))  # Tamaño opcional del gráfico
plt.plot(tiempo, posicion, marker='o', color='purple')  # Color violeta
plt.title('Posición vs Tiempo')  # Título del gráfico
plt.xlabel('Tiempo (s)')  # Etiqueta del eje x
plt.ylabel('Posición (m)')  # Etiqueta del eje y
plt.grid(True)  # Habilitar cuadrícula

# Mostrar el gráfico
plt.show()
```
