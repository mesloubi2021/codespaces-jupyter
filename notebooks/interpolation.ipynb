```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de posición y tiempo
X = np.array([0, 10, 20, 30, 40, 50])
Y = np.array([30, 52, 38, 0, -37, -53])

# Interpolación polinómica de Lagrange
poly = np.poly1d(np.polyfit(X, Y, len(X) - 1))

# Generar puntos para la función interpolada
x_vals = np.linspace(X[0], X[-1], 100)
y_interp = poly(x_vals)

# Graficar la función interpolada
plt.figure(figsize=(5, 3))
plt.plot(X, Y, 'bo', label='Datos originales')  # Puntos originales
plt.plot(x_vals, y_interp, 'g-', label='Interpolación polinómica de Lagrange')  # Función interpolada
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Interpolación Polinómica de Lagrange')
plt.grid(True)
plt.legend()
plt.show()

print(poly)
```
