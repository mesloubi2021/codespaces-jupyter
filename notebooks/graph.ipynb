# Posición del automóvil en varios tiempos

```python
import pandas as pd

# Sample data
data = {'Posición': ['A', 'B', 'C', 'D', 'E', 'F'],
        't(s)': [0, 10, 20, 30, 40, 50],
        'x(m)': [30, 52, 38, 0, -37, -53]}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame as a table
df.style.hide(axis="index")
```

```python
import matplotlib.pyplot as plt

# Plot x(m) vs t(s)
plt.figure(figsize=(5, 3))
plt.plot(df['t(s)'], df['x(m)'], marker='o')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Gráfico de Posición vs Tiempo')
plt.grid(True)
plt.show()
```
